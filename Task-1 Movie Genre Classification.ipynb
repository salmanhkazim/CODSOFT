# Importing necessary libraries

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
```

# Importing and exploring the training data

```python
train_data = pd.read_csv("train_data.txt", sep=':::', names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION'])
display(train_data.head())
print("Training Data Shape:", train_data.shape)
```

# Importing and exploring the testing data

```python
test_data = pd.read_csv("test_data.txt", sep=':::', names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION'])
print(display(test_data.head()))
print("Testing Data Shape:", test_data.shape)
```

# Importing and exploring the solution data for testing

```python
test_solution_data = pd.read_csv("test_data_solution.txt", sep=':::', names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION'])
print(display(test_solution_data.head()))
print("Testing Solution Data Shape:", test_solution_data.shape)
```

# Performing data visualization

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=train_data['GENRE'].value_counts(), y=train_data['GENRE'].value_counts().index, palette='viridis')
plt.title('Number of Movies per Genre')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.show()
```

# Bar Representation of Genre vs Description length

```python
plt.figure(figsize=(15, 10))
sns.barplot(x=train_data['DESCRIPTION'].apply(len), y='GENRE', data=train_data, palette='Set2')
plt.title('Description Length by Genre')
plt.xlabel('Description Length')
plt.ylabel('Genre')
plt.xticks(rotation=45)
plt.show()
```

# Top 10 Most Frequent Genres

```python
top_genres = train_data['GENRE'].value_counts().head(10)
plt.figure(figsize=(15, 8))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='pastel')
plt.title('Top 10 Most Frequent Genres')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.gca().invert_yaxis()
plt.show()
```

# Handling missing values

```python
train_data['DESCRIPTION'].fillna("", inplace=True)
test_data['DESCRIPTION'].fillna("", inplace=True)
```

# Vectorizing the text data

```python
t_v = TfidfVectorizer(stop_words='english', max_features=100000)
X_train = t_v.fit_transform(train_data['DESCRIPTION'])
X_test = t_v.transform(test_data['DESCRIPTION'])
```

# Encoding the labels

```python
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(train_data['GENRE'])
y_test = label_encoder.transform(test_solution_data['GENRE'])
```

# Splitting the data for training and validation

```python
X_train_sub, X_val, y_train_sub, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

# Training a Support Vector Machine (SVM) model

```python
clf = LinearSVC()
clf.fit(X_train_sub, y_train_sub)
```

# Evaluating the model on validation data

```python
y_val_pred = clf.predict(X_val)
print("Validation Accuracy (SVM):", accuracy_score(y_val, y_val_pred))
print("Validation Classification Report (SVM):\n", classification_report(y_val, y_val_pred))
```

# Training a Multinomial Naive Bayes (MNB) model

```python
Mnb_classifier = MultinomialNB()
Mnb_classifier.fit(X_train, y_train)
```

# Predicting on test data using MNB model

```python
Mnb_pred = Mnb_classifier.predict(X_test)
print("Test Accuracy (MNB):", accuracy_score(y_test, Mnb_pred))
print("Test Classification Report (MNB):\n", classification_report(y_test, Mnb_pred))
```

# Training a Logistic Regression model

```python
lr_classifier = LogisticRegression(max_iter=500)
lr_classifier.fit(X_train, y_train)
```

# Predicting on test data using Logistic Regression model

```python
lr_pred = lr_classifier.predict(X_test)
print("Test Accuracy (Logistic Regression):", accuracy_score(y_test, lr_pred))
print("Test Classification Report (Logistic Regression):\n", classification_report(y_test, lr_pred))
```

# Designing a function to predict the genre of a movie based on its description

```python
def predict_movie(description):
    t_v1 = t_v.transform([description])
    pred_label = clf.predict(t_v1)
    return label_encoder.inverse_transform(pred_label)[0]
```

# Testing the prediction function with sample descriptions

```python
sample_descr_for_movie = "A movie where police chase the criminal and shoot him"
print("Predicted Genre for Sample Movie 1:", predict_movie(sample_descr_for_movie))

sample_descr_for_movie1 = "A movie where a person proposes to a girl but she refuses him"
print("Predicted Genre for Sample Movie 2:", predict_movie(sample_descr_for_movie1))
```
